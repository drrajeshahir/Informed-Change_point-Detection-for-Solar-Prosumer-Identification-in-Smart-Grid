```python
#Calculation of size of solar panel for every customers STARTS
df_solar = pd.read_csv("Data/KOL_15mins_final_data-11839.csv")
```

```python
df_solar.head()
```

```python
meterids = pd.DataFrame(df_solar['device_id'].unique())
```

```python
meterids.columns=['meterid']
```

```python
meterids.head()
```

```python
meterid_solarSize = pd.DataFrame(columns=['meterid', 'annual_consumption', 'solar_size'])
```

```python
cuf = 0.1820
allowed_capacity = 0.9
for index, row in meterids.iterrows():
    df_m = df_solar[df_solar['device_id'] == meterids.meterid[index]]
    annual_consu = df_m.values[:,2:98].sum()
    solar_size = (annual_consu/(cuf*9504))*allowed_capacity #13 months data. 396 days * 24 hours = 9504
    meterid_solarSize = meterid_solarSize.append({'meterid':meterids.meterid[index], 'annual_consumption':annual_consu, 'solar_size':solar_size}, ignore_index = True)
```

```python
meterid_solarSize.to_excel("Solar size per meter ID.xlsx", index=False)
#Calculation of size of solar panel for every customers ENDS
```

```python
#Generation of solar average profiles from last 3 years data (2018, 2019, 2020) STARTS
df_kw = pd.read_csv("Solar generation-10.5kw.csv")
```

```python
df_kw.drop('dd-mm only', axis=1, inplace=True)
```

```python
df_kw['date'] = pd.to_datetime(df_kw['date'])
```

```python
#df_1kw['month'] = df_1kw['date'].dt.month
#df_1kw['day'] = df_1kw['date'].dt.day
```

```python
df_kw['date-month'] = pd.to_datetime(df_kw['date']).dt.strftime('%d-%m')
```

```python
df_kw.head()
```

```python
dmlist = list(np.unique(np.array(df_kw['date-month'])))
```

```python
solar_avg = pd.DataFrame(columns=df_kw.columns)
```

```python
solar_avg.drop(['date-month'], axis = 1)
```

```python
for index in range(len(dmlist)):
    day_average = df_kw.iloc[:,1:97][df_kw['date-month'] == dmlist[index]].mean()
    solar_avg = solar_avg.append(day_average, ignore_index=True)
```

```python
solar_avg.to_excel("solar average profile-10.5kw.xlsx", index=False)
```
