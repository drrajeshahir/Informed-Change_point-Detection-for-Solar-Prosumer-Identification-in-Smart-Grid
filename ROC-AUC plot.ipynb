```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
from matplotlib import pyplot
import pandas as pd
from sklearn import metrics
from sklearn import preprocessing
```

```python
font = {'family' : 'times new roman',
        'weight' : 'bold',
        'size'   : 16}

pyplot.rc('font', **font)
pyplot.rcParams["font.weight"] = "bold"
pyplot.rcParams["axes.labelweight"] = "bold"
pyplot.rcParams['figure.figsize'] = 6, 5
```

```python
fpr = [0, 0.9, 0.99, 1.04, 1.05, 1.08, 1.2, 1.88]
#fpr = [0, 0.009011039, 0.009912142, 0.010362694, 0.010475332, 0.01081, 0.01171435, 0.018810543]
```

```python
xmin = min(fpr) 
xmax=max(fpr)
for i, x in enumerate(fpr):
    fpr[i] = (x-xmin) / (xmax-xmin)
print('Normalized List:',fpr)
```

```python
tpr = [0, 85.56, 89.66, 92.83, 95.79, 97.32, 98.24, 99.03]
#tpr = [0, 0.725611858, 0.806618407, 0.888314374, 0.957974509, 0.97932, 0.982419855, 0.990348156]
```

```python
xmin = min(tpr) 
xmax=max(tpr)
for i, x in enumerate(tpr):
    tpr[i] = (x-xmin) / (xmax-xmin)
print('Normalized List:',tpr)
```

```python
auc(fpr, tpr)
```

```python
# plot no skill roc curve
pyplot.plot([0, 1], [0, 1], linestyle='--')
# calculate roc curve for model
#fpr, tpr, _ = roc_curve(testy, pos_probs)
# plot model roc curve
pyplot.plot(fpr, tpr, marker='*', label='ROC (AUC=0.76)', lw = 2)
#addlabels(fpr, tpr)
# axis labels
pyplot.xlabel('False Positive Rate')
pyplot.ylabel('True Positive Rate')
pyplot.xlim([0, 1])
# show the legend
pyplot.legend()
# show the plot
pyplot.savefig("Results/ROC curve_600_5x5.tiff", dpi=600, bbox_inches='tight')
pyplot.show()
```
